```python
import pandas as pd
import datetime as dt
import matplotlib.pyplot as plt
import numpy as np
```

```python
df_1 = pd.read_json('../results/test_run_3/output_1.json')
df_2 = pd.read_json('../results/test_run_3/output_2.json')
df_3 = pd.read_json('../results/test_run_3/output_3.json')
df_4 = pd.read_json('../results/test_run_3/output_4.json')
df_5 = pd.read_json('../results/test_run_3/output_5.json')
```

```python
df_1[df_1['msg'] == 'Stats'][['time','total']]
```

```python
df_1['timestamp'] = pd.to_datetime(df_1['timestamp'])
df_2['timestamp'] = pd.to_datetime(df_2['timestamp'])
df_3['timestamp'] = pd.to_datetime(df_3['timestamp'])
df_4['timestamp'] = pd.to_datetime(df_4['timestamp'])
df_5['timestamp'] = pd.to_datetime(df_5['timestamp'])
```

```python
df_1[df_1["service"] == 'listener']
```

```python
df_2[df_2["service"] == 'listener']
```

```python
df_3[df_3["service"] == 'listener']
```

```python
df_1_lis_time_delta = df_1[df_1["service"] == 'listener'][['timestamp', 'hash']]
df_2_lis_time_delta = df_2[df_2["service"] == 'listener'][['timestamp', 'hash']]
df_3_lis_time_delta = df_3[df_3["service"] == 'listener'][['timestamp', 'hash']]
df_4_lis_time_delta = df_4[df_4["service"] == 'listener'][['timestamp', 'hash']]
df_5_lis_time_delta = df_5[df_5["service"] == 'listener'][['timestamp', 'hash']]
```

```python
combined_time_delta = df_1_lis_time_delta.set_index('hash').join(df_2_lis_time_delta.set_index('hash'), how='outer', lsuffix='1', rsuffix='2').join(df_3_lis_time_delta.set_index('hash'),how='outer', lsuffix='3', rsuffix='4').join(df_4_lis_time_delta.set_index('hash'),how='outer', lsuffix='5', rsuffix='6').join(df_5_lis_time_delta.set_index('hash'),how='outer', lsuffix='7', rsuffix='8')
```

```python
combined_time_delta['max'] = combined_time_delta.max(axis=1)
combined_time_delta['min'] = combined_time_delta.min(axis=1)
combined_time_delta['delta'] = combined_time_delta['max'] - combined_time_delta['min']
combined_time_delta['delta_s'] = combined_time_delta['delta'].to_numpy() / np.timedelta64(1, 's')
```

```python
combined_time_delta['delta_s']
```

```python
df_1_lis_txs = df_1[df_1["service"] == 'listener'][['txs', 'hash']] 
df_2_lis_txs = df_2[df_2["service"] == 'listener'][['txs', 'hash']] 
df_3_lis_txs = df_3[df_3["service"] == 'listener'][['txs', 'hash']] 
df_4_lis_txs = df_4[df_4["service"] == 'listener'][['txs', 'hash']] 
df_5_lis_txs = df_5[df_5["service"] == 'listener'][['txs', 'hash']] 
```

```python
combined_txs = df_1_lis_txs.set_index('hash').join(df_2_lis_txs.set_index('hash'), how='outer', lsuffix='1', rsuffix='2').join(df_3_lis_txs.set_index('hash'),how='outer', lsuffix='3', rsuffix='4').join(df_4_lis_txs.set_index('hash'),how='outer', lsuffix='5', rsuffix='6').join(df_5_lis_txs.set_index('hash'),how='outer', lsuffix='7', rsuffix='8')
```

```python
combined_txs.max(axis=1)
```

```python
df_1_lis_size = df_1[df_1["service"] == 'listener'][['size', 'hash']] 
df_2_lis_size = df_2[df_2["service"] == 'listener'][['size', 'hash']] 
df_3_lis_size = df_3[df_3["service"] == 'listener'][['size', 'hash']] 
df_4_lis_size = df_4[df_4["service"] == 'listener'][['size', 'hash']] 
df_5_lis_size = df_5[df_5["service"] == 'listener'][['size', 'hash']] 
```

```python
combined_size = df_1_lis_size.set_index('hash').join(df_2_lis_size.set_index('hash'), how='outer', lsuffix='1', rsuffix='2').join(df_3_lis_size.set_index('hash'),how='outer', lsuffix='3', rsuffix='4').join(df_4_lis_size.set_index('hash'),how='outer', lsuffix='5', rsuffix='6').join(df_5_lis_size.set_index('hash'),how='outer', lsuffix='7', rsuffix='8')
```

```python
combined_size.max(axis=1)
```

```python
combined_new = combined_time_delta[['min', 'delta_s']].join(combined_txs.max(axis=1).rename('txs')).join(combined_size.max(axis=1).rename('size')).sort_values(by='min', ascending=True)
```

```python
combined_new['min'].min()
```

```python
time_delta = combined_new['min'] - combined_new['min'].min()
```

```python
combined_new['ts'] = time_delta.to_numpy() / np.timedelta64(1, 's')
```

```python
combined_final = combined_new.set_index('ts').drop(columns=['min'])
combined_final
```

```python
combined_final['txs'].index
```

```python
plt.figure(figsize=(10,5))
plt.scatter(combined_final['txs'].index, combined_final['txs'])
plt.xlabel('time delta [s]')
plt.ylabel('nr of txs')
plt.legend()
plt.show()
```

```python
plt.figure(figsize=(10,5))
plt.scatter(combined_final['size'].index, combined_final['size'])
plt.ylabel('block size [B]')
plt.xlabel('time delta [s]')
plt.legend()
plt.show()
```

```python
plt.figure(figsize=(10,5))
plt.scatter(combined_final['delta_s'].index, combined_final['delta_s'])
plt.ylabel('block consensus time delta [s]')
plt.xlabel('time [s]')
plt.legend()
plt.show()
```
